```python
def preprocess_duration(x):
    if 'h' not in x:
        x = '0h ' + x
        
    elif 'm' not in x:
        x= x + ' 0m'
        
    return x
```

#### 15. How to deal with Outlier

```python
data['Price'].describe()
```

```python
q1 = 5277.000000
q2 = 8372.000000
q3 = 12373.000000
iqr = q3-q1

maxi = q3 + 1.5*iqr
mini = q1 - 1.5*iqr
```

```python
maxi
```

```python
outliers = [price for price in data['Price'] if price>maxi or price<mini]
```

```python
len(outliers)
```

* Mean is affected by outliers in data, median is not affected
* Replace outtliers with median price

```python
#  Suppose Domain Expertise given threshold as 35k
```

```python
data['Price'] = np.where(data['Price']>=35000, data['Price'].median(), data['Price'])
```

```python
plot(data, 'Price')
```

```python
# Now skewness has decreased, due to outliers removal
```

#### 18. How to define your Evaluation metric......

```python
# mape --> mean absolute percentage error
```

```python
def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs( (y_true - y_pred) / y_true )) *100
```

```python
mape(y_test, y_pred)
```

#### 19. How to automate Machine Learning Pipeline

```python
def predict(ml_model):

    model = ml_model.fit(X_train, y_train)
    print(f'Training score {model.score(X_train, y_train)}')
    y_prediction = model.predict(X_test)
    print(f'predictions are : {y_prediction}')
    print('\n')
    r2_score = metrics.r2_score(y_test, y_prediction)
    print(f'r2 score: {r2_score}')
    print(f'MAE : {metrics.mean_absolute_error(y_test, y_prediction)}')
    print(f'MSE : {metrics.mean_squared_error(y_test, y_prediction)}')
    print(f'RMSE : {np.sqrt(metrics.mean_squared_error(y_test, y_prediction))}')
    print(f'MAPE : {mape(y_test, y_prediction)}')
    sns.distplot(y_test-y_prediction)
    
```

```python
predict(RandomForestRegressor())
```

#### 20. How to hypertune model

```python
reg_rf = RandomForestRegressor()
```

```python
np.linspace(start=100, stop=1200, num=6)
```

```python
random_grid = {
    
    'n_estimators' : [int(x) for x in np.linspace(start=100, stop=1200, num=6)],
    'max_features' : ['auto', 'sqrt'],
    'max_depth' : [int(x) for x in np.linspace(start=5, stop=30, num=4)],
    'min_samples_split' : [5, 10, 15, 100] 
}
```

```python
random_grid           # this dictionary is hyper parameter space
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf_random = RandomizedSearchCV(estimator=reg_rf, param_distributions=random_grid, cv=3, n_jobs=-1, verbose=2)
```

```python
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_estimator_
```
